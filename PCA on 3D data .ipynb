```python
import numpy as np 
import matplotlib.pyplot as plt 
from sklearn.decomposition import PCA
```

```python
np.random.seed(2343243)
```

```python
mean_vec1 = np.array([0,0,0])
conv_mat1 = np.array([[1,0,0],[0,1,0],[0,0,1]])
class1 = np.random.multivariate_normal(mean_vec1, conv_mat1, 100)
```

```python
mean_vec2 = np.array([1,1,1])
conv_mat2 = np.array([[1,0,0], [0,1,0], [0,0,1]])
class2 = np.random.multivariate_normal(mean_vec2, conv_mat2, 100)
```

```python
from matplotlib import pyplot as plt 
from mpl_toolkits.mplot3d import Axes3D, proj3d 

fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(111, projection = '3d')
ax.plot(class1[:,0], class1[:,1], class1[:,2], 'o')
ax.plot(class2[:,0], class2[:,1], class2[:,2], '^')
plt.show()
```

```python
all_data = np.concatenate((class1,class2))
all_data.shape
```

```python
pca = PCA(n_components=2)
transformed_data = pca.fit_transform(all_data)
```

```python
pca.components_
```

```python
plt.scatter(transformed_data[:100,0],transformed_data[:100,1])
plt.scatter(transformed_data[100:,0],transformed_data[100:,1])
plt.show()
```

```python
X_inverse = pca.inverse_transform(transformed_data)
```

This transformed data lines in a plane. As we have earlier converted the 3d data to 2d data.

```python
from matplotlib import pyplot as plt 
from mpl_toolkits.mplot3d import Axes3D, proj3d 
fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(111, projection = '3d')
ax.plot(X_inverse[:,0], X_inverse[:,1], X_inverse[:,2], 'o')
plt.show()
```
